```python
file = open("build/timestamps_RBTree.txt", "r")
timestamps_RBTree = file.readlines()
file.close()

file = open("build/timestamps_BITree.txt", "r")
timestamps_BITree = file.readlines()
file.close()

file = open("build/timestamps_HTable.txt", "r")
timestamps_HTable = file.readlines()
file.close()

file = open("build/timestamps_MultiMap.txt", "r")
timestamps_MultiMap = file.readlines()
file.close()
```

```python
sizes = [10, 100, 500, 1000, 2000, 10000, 15000, 30000, 50000, 100000]
```

```python
import matplotlib.pyplot as plt
```

```python
print(len(timestamps_HTable))
print(len(sizes))
```

```python
plt.title('Зависимости времени поиска от размерности массива')
plt.xlabel('Размерность массива')
plt.ylabel('Время, наносекунды')
plt.plot(sizes, timestamps_BITree, color='r', marker='o')
plt.plot(sizes, timestamps_HTable, color='b', marker='o')
plt.plot(sizes, timestamps_RBTree, color='g', marker='o')
plt.plot(sizes, timestamps_MultiMap, color='y', marker='o')
plt.legend(['BITree', 'HTable', 'RBTree', 'MultiMap'])
plt.show()
```

```python
file = open("build/collision.txt", 'r')
collisions = file.readlines()
file.close()
```

```python
plt.title('График зависимости числа коллизий от размера массива')
plt.xlabel('Размерность массива')
plt.ylabel('Количество колизий')
plt.plot(sizes, collisions, marker='o')
plt.show()
```
